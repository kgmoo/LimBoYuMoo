```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# First we need to get back the file produced by GEOParse
rawDataFrameGSE10245 = pd.read_csv(r'~/Documents/BIOINF575/GroupProject3/LimBoYuMoo/rawDataFrameGSE10245.csv', index_col=0)
print(rawDataFrameGSE10245)
rawDataFrameGSE10245_T = rawDataFrameGSE10245.T
print(rawDataFrameGSE10245_T)
```

```python
# Then we need to generate a K-means object using the appropriate number (k) of means
kmeansInit = KMeans(n_clusters = 2)
# We give the kmeans object the data in question to generate the cluster points
kmeansInit.fit(rawDataFrameGSE10245_T)
# Finally we ask it to predict the data used to train it and generate the cluster codes for each sample
GSE10245Means = kmeansInit.predict(rawDataFrameGSE10245_T)
# Let's attach these values to a Series labeled with the sample names for the benefit of those who follow
GSE10245ClusterCodes = pd.Series(GSE10245Means, index = rawDataFrameGSE10245_T.index)
GSE10245ClusterCodes.to_csv(r'~/Documents/BIOINF575/GroupProject3/LimBoYuMoo/clusterCodesGSE10245.csv')
```
