```python
metadata_dict = {}
for gsm_name, gsm in rawDataGSE10245.gsms.items():
    for key, value in gsm.metadata.items():
        if key == 'characteristics_ch1':
            split_value = value[0].split(':')
            metadata_dict[gsm_name] = split_value[1]
        else:
            continue
subtype_series =  pd.Series(metadata_dict)
```

```python
subtype_df= subtype_series.to_frame('subtype')
```

```python
annotated_df = dataframe.join(subtype_df, on='sample_label')
```

```python
def change_subtype_label(old_label: str, new_label: str, df) -> pd.DataFrame:
    
    current_labels = df.subtype.unique()
    
    if not old_label in current_labels:
        raise ValueError("inputed old_label is not a current subtype label! current labels are: ", current_labels )
    
    changed_df = df.replace(old_label, new_label, regex=True)
    return changed_df
```
